# Medical Insurance Prediction

This project explores a dataset that provides information about 100,000 individuals including their demographics, socioeconomic status, health conditions, lifestyle factors, insurance plans, and medical expenditures. The purpose of this project is to develop a machine learning model that can predict the annual medical cost of a patient, based on their medical condition and history. 
The dataset was cleaned and preprocessed for ease of analysis and is available under an open license for research and data analysis purposes. The dataset is provided by a user from Kaggle (https://www.kaggle.com/datasets/mohankrishnathalla/medical-insurance-cost-prediction). We will be using Python to analyze the dataset and a series of visualisations such as pandas, seaborn, matplotlib and SciPy libraries to answer the following questions:

1. Is there a significant difference between different weight classes in medical costs?
2. How does age affect medical costs?
3. How do the feature variables correlate with the target variable?

## Data Cleaning and Preparation

```python
# Dropping columns
df.drop(columns = {'person_id'}, inplace = True)

# Filling null values
df.alcohol_freq.fillna(df.alcohol_freq.mode()[0], inplace = True)

# Logging medical cost
df['log_cost'] = np.log(df.annual_medical_cost)
```

```python
# Converting data type of columns
bin_cols = ['hypertension', 'diabetes', 'copd', 'cancer_history', 'kidney_disease', 'liver_disease', 'arthritis', 'mental_health', 
            'systolic_bp', 'diastolic_bp', 'ldl', 'hba1c', 'is_high_risk', 'had_major_procedure']
for col in bin_cols:
    df[col].replace({1:True, 0:False}, inplace = True)
```

```python
# Binning bmi
def bmi_category(bmi):
    if bmi < 18.5: return 'Underweight'
    elif bmi < 25: return 'Normal'
    elif bmi < 30: return 'Overweight'
    else: return 'Obese'
df['bmi_category'] = df['bmi'].apply(bmi_category)

# Binning age
bins = [0, 18, 30, 45, 60, 100]
labels = ['Child', 'Young Adult', 'Adult', 'Middle Aged', 'Senior']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels)
```

```python
# Replacing value in education
df['education'].replace({'Some College':'College'}, inplace = True)

# Dropping duplicates
df.drop_duplicates(inplace=True)
```

### Data Question 1: Is there a significant difference in medical costs between different weight classes?

```python
# Distribution of BMI
sns.histplot(data = df, x = 'bmi', bins = 50)
plt.title('Distribution of BMI')
plt.show()
```

The normal range of BMI is between 18.5 and 24.9. The distribution peaks around 28 which signifies that there are more people who are considered overweight. 

```python
# First and third quartile of annual medical cost
Q1 = df['annual_medical_cost'].quantile(0.25)
Q3 = df['annual_medical_cost'].quantile(0.75)
print('The first quartile of annual medical cost is $'+ str(round(Q1,2)))
print('The third quartile of annual medical cost is $'+ str(round(Q3,2)))

# Creating dataframe
df_q1 = df[df.annual_medical_cost < Q1]
df_q3 = df[df.annual_medical_cost > Q3]
```

```python
# Distribution of weight class 
plt.figure(figsize = (9,4), facecolor = 'aliceblue')
plt.subplots_adjust(wspace = 0.3)
plt.suptitle('Distribution of Weight Class', y = 1.02)

# Low medical cost
plt.subplot(1,2,1)
sns.countplot(data = df_q1, x = 'bmi_category', palette = 'mako_r', order = ['Underweight', 'Normal', 'Overweight', 'Obese'])
plt.title('Low Cost')
plt.xlabel('')
plt.ylabel('Frequency')
plt.xticks(rotation = 45)
plt.ylim(0, 10000)

# High medical cost
plt.subplot(1,2,2)
sns.countplot(data = df_q3, x = 'bmi_category', palette = 'mako_r', order = ['Underweight', 'Normal', 'Overweight', 'Obese'])
plt.title('High Cost')
plt.xlabel('')
plt.ylabel('Frequency')
plt.xticks(rotation = 45)
plt.ylim(0, 10000)
plt.show()
```

It seems that weight does play a factor in annual medical cost. There are more obese people who have a higher medical cost. People who are underweight or have normal weight are more common at a lower cost. 

```python
# OLS
ols_formula = 'log_cost ~ C(bmi_category)'
model = ols(formula = ols_formula, data = df).fit()
model.summary()
```

Since the p-value is below our significance threshold of 0.05, we can reject the null hypothesis and accept the alternative hypothesis which states that there is a significant difference in annual medical cost between weight classes. 

```python
# Performing pairwise comparisons with different weight classes
tukey_oneway = pairwise_tukeyhsd(endog = df['log_cost'], groups = df['bmi_category'])
tukey_oneway.summary()
```

The test above shows the comparison between all the weight classes. The `reject` column displays `True` for all rows which means that there is significant difference among all classes.

### Data Question 3: How do the feature variables correlate with the target variable?

```python
# Creating correlation matrix
num_cols = ['age', 'income', 'household_size', 'dependents', 'bmi', 'visits_last_year', 'hospitalizations_last_3yrs', 'days_hospitalized_last_3yrs', 
            'medication_count', 'risk_score', 'annual_medical_cost', 'claims_count', 'avg_claim_amount', 'total_claims_paid']
corr_matrix = df[num_cols].corr(method = 'pearson')

# Plotting heatmap for correlation between independent variables
plt.figure(figsize = (14,6))
sns.heatmap(corr_matrix, annot = True, cmap = 'Blues')
plt.title('Correlation Matrix of Feature Variables')
plt.show()
```

There are a few variables with high multicollinearity such as household_size with dependents and hospitalizations_last_3yrs with days_hospitalized_last_3yrs. We will remove one variable each from pairs with high multicollinearity to reduce noise when training our model. 

```python
# Plotting heatmap between independent and dependent variable
corr_target = corr_matrix[['annual_medical_cost']].drop(labels = 'annual_medical_cost')

plt.figure(figsize = (5,4))
sns.heatmap(corr_target, annot = True, cmap = 'Blues')
plt.title('Correlation Between Feature and Target Variables')
plt.show()
```

Using Pearson's correlation, there are two variables with high correlation with the target variable; avg_claim_amount and total_claims_paid with a correlation of 0.63 and 0.74 respectively.

## Model Selection

```python
def make_results(model_name: str, model_object, metric: str):
    '''
    Arguments:
        model_name (string): name for the model in output table
        model_object: a fit GridSearchCV object
        metric (string): 'rmse', 'mae', or 'r2'

    Returns:
        pandas df with RMSE, MAE, and RÂ² for the best model based on the chosen metric
    '''
    # Get CV results
    best_idx = model_object.best_index_
    cv_results = model_object.cv_results_

    # CV RMSE (average over folds for best params)
    cv_rmse = np.sqrt(-cv_results['mean_test_neg_mean_squared_error'][best_idx])
    r2 = cv_results['mean_test_r2'][best_idx]
    
    # Predict on test set
    y_pred_test = model_object.best_estimator_.predict(x_test)

    # Test RMSE
    test_rmse = mean_squared_error(y_test, y_pred_test)
    test_r2 = r2_score(y_test, y_pred_test)

    dataframe = pd.DataFrame({
        'model': [model_name],
        'RMSE': [cv_rmse],
        'R2': [r2] })
        
    result = pd.concat([dataframe],axis = 0)
          
    return result
```

### Splitting dataset

```python
# Feature and taget variables
X = df.drop(columns = {'annual_medical_cost', 'annual_premium', 'monthly_premium', 'log_cost', }).copy()
y = df.log_cost

# Splitting data into train, validation and test data
x_tr, x_test, y_tr, y_test = train_test_split(X, y, train_size = 0.75, random_state = 42)
x_train, x_val, y_train, y_val = train_test_split(x_tr, y_tr, test_size = 0.2, random_state = 42)
```

### Random Forest Regressor

```python
# Instantiate Random Forest Regressor
rf = RandomForestRegressor(random_state = 42)

# Defining RF hyperparameters for GridSearchCV 
cv_params = {'max_depth': [None],
             'max_features': [None],
             'max_samples': [1],
             'min_samples_leaf': [2],
             'min_samples_split': [2],
             'n_estimators': [200, 300],
             }

# Scoring metrics
scoring = ['r2', 'neg_mean_squared_error', 'neg_mean_absolute_error']

# Instantiate GridSearchCV for RF
rf_cv = GridSearchCV(rf, cv_params, scoring = scoring, cv = 3, refit = 'r2')

# Fitting model with GridSearchCV
rf_cv.fit(x_train, y_train)
```

```python
# Calling make_results function
result = make_results('RF', rf_cv, 'r2')
result
```
